```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('aggs').getOrCreate()
```

```
df =spark.read.csv('/FileStore/tables/sales_info.csv', inferSchema=True,header=True)
```

```
df.printSchema()
```

```
df.groupBy(df['Company'])
```

```
df.groupBy(df['Company']).mean().show()
```

```
df.groupBy(df['Company']).sum().show()
```

```
df.groupBy(df['Company']).count().show()
```

```
## Aggerigate functions----------Note agg() methos take dictonary as an argument-------------------------
df.agg({'Sales' : 'max'}).show()
```

```
group_data = df.groupBy(df['Company'])
```

```
group_data.agg({'Sales' : 'max'}).show()
```

```
from pyspark.sql.functions import countDistinct,avg,stddev
```

```
df.select(countDistinct(df['Sales'])).show()
```

```
df.select(avg(df['Sales']).alias('Average Sales')).show()
```

```
result_std = df.select(stddev(df['Sales']).alias('Standars deviation'))
result_std.show()
```

```
from pyspark.sql.functions import format_number

result_std.select(format_number('Standars deviation', 2)).show()
```

```
# OrderBy and Sorting
#df.show()
df.orderBy(df['Sales']).show()
```

```
# Decesnding Order
df.orderBy(df['Sales'].desc()).show()
```
