```
# Defining own scema , beacause json files doesn't support inferschema
from pyspark.sql.types import (StringType,StructField,
                               IntegerType,StructType)
```

```
data_schema = [StructField('age',IntegerType(), True),
              StructField('name',StringType(),True)]
```

```
final_struc = StructType(fields=data_schema)
```

```
df =spark.read.json('/FileStore/tables/people.json', schema=final_struc)
```

```
df.printSchema()
```

```
df.show()
```

```
type(df['age'])
```

```
# For grabing a column in dataframe we have to use select() method
df.select('age').show()
```

```
df.head(2)
```

```
# for selecting multiple columns , we have to pass a list in select ()
df.select(['age', 'name'])
```

```
# Adding new Columns in dataframe
df.withColumn('new_age', df['age']*2).show()
```

```
# REnaiming an exsitinf column
df.withColumnRenamed('age', 'new_age').show()
```

```
# Through SQL
df.createOrReplaceTempView('people')
```

```
result = spark.sql("SELECT * FROM people")
```

```
result.show()
```

```
## Basics Operation in Spark------------------------------------
df1 = spark.read.csv('/FileStore/tables/appl_stock.csv', inferSchema=True,header=True)
```

```
df1.printSchema()
```

```
df1.head(3)[0]
```

```
# Close less than 500
#df1.filter("Close < 500").show()

# With Dataframe own operator
df1.filter(df1['Close'] < 500).show()
```

```
#df1.filter("Close < 500").select(['Open','Date']).show()

df1.filter(df1['Close'] < 500).select(['Volume']).show()
```

```
# Filtering based on multiple conditions
df1.filter( (df1['Open'] > 200) & (df1['Close'] < 200) ).show()
```

```
df1.filter(df1['Low'] == 197.16).show()
```

```
result1 = df1.filter(df1['Low'] == 197.16).collect()
```
