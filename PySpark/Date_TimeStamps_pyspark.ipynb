```
from pyspark.sql import  SparkSession
```

```
spark = SparkSession.builder.appName('dates').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/appl_stock.csv',inferSchema=True,header=True)
```

```
df.select(['Date','Open']).show()
```

```
from pyspark.sql.functions import (dayofmonth,
                                   dayofyear,
                                   hour,
                                   month,
                                   year,
                                   weekofyear,format_number,date_format)
```

```
df.select(hour(df['Date'])).show()
```

```
#df.select(year(df['Date'])).show()
new_df = df.withColumn("Year",year(df['Date']))
```

```
#new_df.groupBy(new_df['year']).mean().show()

# Average closing price per year.
result = new_df.groupBy(new_df['year']).mean().select(["Year","avg(Close)"])
```

```
new = result.withColumnRenamed("avg(Close)", "Average closing price")
```

```
new.select(['Year',format_number('Average closing price', 2).alias("Avg Close")]).show()
```
